```python
import folium
import pandas as pd
```

```python
dat = pd.read_csv (r'Posicion_Tiendas_Guanajuato.csv')
```

```python
dat_no_NA = dat.dropna()
```

```python
m = folium.Map(
    location=[21.123, -101.682],
    zoom_start=12,
    tiles='Stamen Terrain'
)
```

```python
dat_no_NA.apply(lambda row:folium.CircleMarker(location=[row["latitud"], row["longitud"]], 
                                              radius=3, popup=row['cadenaComercial'])
                                             .add_to(m), axis=1)
```

```python
m.save("Tiendas_Leon.html")
```
